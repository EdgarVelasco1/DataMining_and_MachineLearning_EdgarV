# Crear un arreglo de 10  numeros enteros aleatorios entre el 1 al 100, y despues calcular la media y la desviacion estandar del arreglo

```r
numeros<-sample(1:100,10) 
numeros   
print('media')
mean(numeros)   
print("desviacion estandar",)
sd(numeros)
```

# Crear un arreglo bidimensional de 3 filas y 4 columnas con valores enteros aleatorios, luego calcular la suma de cada fila y cada columna

```r
numeros<-sample(1:100,12)
matriz<-array(c(numeros),dim=c(3,4,1))
print('matriz')
print(matriz)
print('suma de las filas')
print(rowSums(matriz))
print('suma de las columnas')
print(colSums(matriz))
```

# Crear un Dataframe con tres columnas: "Nombre", "Edad", y "Ciudad". Añadir al menos 5 filas de datos para el DataFrame. luego, filtrar el DataFrame para solo incluir filas donde la edad de las personas es mayor o igual que 30.

```r
df<-data.frame(
                "Nombre" = as.character(c('Erick','Anna','Ahoshino','Khang','Marissa')),
                "Edad" = c(19,21,25,43,45),
                "Ciudad" = as.character(c('New York','Pripiat','Saitama','Bombay','Napoles'))
)

df[df$Edad>=30,]
```

# Cargar un archivo CSV dentro de un DataFrame y luego calcular la media, mediana y la moda de una de las columnas en el dataframe.

```r
calcular <- read.csv('/kaggle/input/calificacionesr/calificaciones.csv',header=TRUE,sep=";")
calcular
print("la media de c1 es:")
valor <- calcular$calificaciones1
mean(valor)
print("la mediana de c1 es:")
median(valor)

moda <- function(x) {
   return(as.numeric(names(which.max(table(x)))))
}
print("la moda de c1 es:")
moda(valor)
```
