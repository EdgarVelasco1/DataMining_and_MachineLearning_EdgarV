# Crear un arreglo bidimensional de 3 filas y 4 columnas con valores enteros aleatorios, luego calcular la suma de cada fila y cada columna

```r
numeros<-sample(1:100,12)
matriz<-array(c(numeros),dim=c(3,4,1))
print('matriz')
print(matriz)
print('suma de las filas')
print(rowSums(matriz))
print('suma de las columnas')
print(colSums(matriz))
```

### w) Por cada mes, obtener la temperatura maxima y mínima registrada en grados Celsius

```r
library(dplyr)
df <- read.csv('/kaggle/input/new-york-air-quality/airquality.csv',header=TRUE,sep=",")
df1 <- mutate(df, TempC = (Temp-32)*5/9)

Temperatura <- df1 %>%
  group_by(Month) %>%
  summarize(temp_max = max(TempC),
            temp_min = min(TempC))
Temperatura
```

### x) Obtener el promedio de los siguientes parámetros por mes: temperatura en grados Celsius, viento, radiación solar y ozono.

```r
library(dplyr)
df <- read.csv('/kaggle/input/new-york-air-quality/airquality.csv',header=TRUE,sep=",")
df1 <- mutate(df, TempC = (Temp-32)*5/9)
df1$Ozone[is.na(df1$Ozone)] <- 0
df1$Solar.R[is.na(df1$Solar.R)] <- 0

Promedios <- df1 %>%
  group_by(Month) %>%
  summarize(p_temperatura = mean(TempC),
            p_viento = mean(Wind),
            p_radiacion_solar = mean(as.numeric(Solar.R)),
            p_ozono = mean(as.numeric(Ozone)))
Promedios
```
